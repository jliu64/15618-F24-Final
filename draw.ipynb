```python
import matplotlib.pyplot as plt
import numpy as np

# Data for OpenMP H = 2
sequential_time_h2 = 49.7859111530  # Sequential execution time for H = 2
threads = [2, 4, 8, 16, 32, 64, 128]  # Number of threads
openmp_times_h2 = [
    27.8243809200,  # 2 threads
    15.5873903160,  # 4 threads
    9.8541400240,   # 8 threads
    6.6902793120,   # 16 threads
    6.9242295350,   # 32 threads
    6.6816680800,   # 64 threads
    7.2009763110    # 128 threads
]

# Data for OpenMP H = 3
sequential_time_h3 = 125.2151988370  # Sequential execution time for H = 3
openmp_times_h3 = [
    68.5046913950,  # 2 threads
    37.2623587030,  # 4 threads
    22.4870735480,  # 8 threads
    15.1175001660,  # 16 threads
    12.7560851610,  # 32 threads
    13.9543705640,  # 64 threads
    14.0901227850   # 128 threads
]

# Calculate Speed-Up
speed_up_h2 = [sequential_time_h2 / t for t in openmp_times_h2]
speed_up_h3 = [sequential_time_h3 / t for t in openmp_times_h3]

# Create the plot
plt.figure(figsize=(10, 6))

# Plot for H = 2
x_ticks_h2 = np.arange(len(threads))
plt.plot(x_ticks_h2, speed_up_h2, marker='o', label='OpenMP Speed-Up (H = 2)', color='blue')
for x, y in zip(x_ticks_h2, speed_up_h2):
    plt.text(x, y + 0.1, f'{y:.2f}x', fontsize=9, ha='center', va='bottom')

# Plot for H = 3
x_ticks_h3 = np.arange(len(threads))
plt.plot(x_ticks_h3, speed_up_h3, marker='o', label='OpenMP Speed-Up (H = 3)', color='green')
for x, y in zip(x_ticks_h3, speed_up_h3):
    plt.text(x, y + 0.1, f'{y:.2f}x', fontsize=9, ha='center', va='bottom')

# Add titles and labels
plt.title("PSC: OpenMP Speed-Up vs Number of Threads (H = 2 and H = 3)")
plt.xlabel("Number of Threads")
plt.ylabel("Speed-Up (Times Faster than Sequential)")

# Set custom x-axis labels
plt.xticks(x_ticks_h2, [str(t) for t in threads])

# Set y-axis limits
plt.ylim(0, max(max(speed_up_h2), max(speed_up_h3)) + 1)

# Add grid, legend, and layout
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
